## IMPORT PACKAGES ##

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns ; sns.set()
from matplotlib.pyplot import style ; style.use('ggplot')
```

### TRAVERSING DATASET ###

```python
df = pd.read_csv('voice.csv')
df.head()
```

## NULL VALUES ##

```python
df.isna().sum()/len(df)
```

```python
df.label=df.label.map({'male':0,'female':1})
```

## CORRELATION ##

```python
df.corr('spearman')
```

```python
plt.figure(figsize=(24,22))
sns.heatmap(df.corr(),annot=True,cmap = 'RdYlGn')
plt.show()
```

```python
x = df[['meanfreq','sd','median','Q25','meanfun','IQR']].values
y = df[['label']].values
```

### DATA DIMENSION REDUCTION ###

```python
from sklearn.decomposition import PCA
pca = PCA(2)
x_pca = pca.fit_transform(x,y)
```

```python
plt.figure(figsize=(24,22))
sns.scatterplot(x=x_pca[:,0],y=x_pca[:,1],hue=y.reshape(1,-1)[0])
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(2)
x_tsne=tsne.fit_transform(x,y)
```

```python
plt.figure(figsize=(24,22))
sns.scatterplot(x=x_tsne[:,0],y=x_tsne[:,1],hue=y.reshape(1,-1)[0])
plt.show()
```

### SPLITING DATASET AND MODELLING ###

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x,y)
```

```python
from sklearn.svm import SVC
svm = SVC(gamma=0.001)
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
rf = RFC()
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier as GBC
gb = GBC()
gb.fit(x_train,y_train)
gb.score(x_test,y_test)
```

```python
from xgboost import XGBClassifier as XG
xgb = XG()
xgb.fit(x_train,y_train)
xgb.score(x_test,y_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier as ADC
ad = ADC()
ad.fit(x_train,y_train)
ad.score(x_test,y_test)
```

```python
from sklearn.ensemble import ExtraTreesClassifier as ETC
et = ETC()
et.fit(x_train,y_train)
et.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier as DT
dt = DT()
dt.fit(x_train,y_train)
dt.score(x_test,y_test)
```

```python
xt_train,xt_test,yt_train,yt_test = tts(x_tsne,y)
```

```python
svm2 = SVC(gamma=0.001)
svm2.fit(xt_train,yt_train)
svm2.score(xt_test,yt_test)
```

```python
rf2 = RFC(n_estimators=100)
rf2.fit(xt_train,yt_train)
rf2.score(xt_test,yt_test)
```

```python
gb2 = GBC(learning_rate=0.01)
gb2.fit(xt_train,yt_train)
gb2.score(xt_test,yt_test)
```
